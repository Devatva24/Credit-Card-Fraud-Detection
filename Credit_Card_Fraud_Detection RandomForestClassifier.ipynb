This dataset is highly unbalanced  
Label 1 = faudulent transaction  
Label 0 = Normal transaction

```python
# we need to separate the fraudulent and normal transactions
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
fraud.Amount.describe()
```

###Under-Sampling Data  
Since the dataset is highly unbalanced hence we balance it by using under sampling method where we build a sample dataset containing similar distribution of normal and fraud transaction

```python
legit_sample = legit.sample(n=38) # we take random 38 samples from legit and we train the machine with the same
```

Concatenate the two datasets

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
new_dataset
```

```python
new_dataset.head()
```

```python
new_dataset["Class"].value_counts()
```

```python
new_dataset.groupby("Class").mean()
```

```python
 X = new_dataset.drop(columns="Class",axis = 1)
 y = new_dataset["Class"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=2)
# stratify takes care of equal distribution of y label in train and test split
# random state ensure reproduction of code as it is
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

###Model Training

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(class_weight='balanced', n_estimators=100, random_state=42)
```

```python
model.fit(X_train, y_train)
```

###Accuracy Score

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, y_train)
```

```python
print("Accuracy Score on Training dataset : ",training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, y_test)
```

```python
print("Accuracy Score on Test dataset : ",test_data_accuracy)
```
